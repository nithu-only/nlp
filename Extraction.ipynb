```python
#manually
import numpy as np
def tf(a,z):
    n=0
    doc=z.split(' ')
    for i in doc:
        if i==a:
            n+=1
    return(n/len(doc))
```

```python
tf('hits',z[0])
```

```python
def idf(a,data):
    n=0
    for i in data:
        if a in i:
            n+=1
    return(np.log(len(data)/n))
```

```python
idf('hits',z)
```
